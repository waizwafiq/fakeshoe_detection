```python
import cv2
import os
import seaborn as sns
import matplotlib.pyplot as plt

path = f'./dataset'
dataset_auth = os.listdir(f'{path}/auth')
dataset_fake = os.listdir(f'{path}/fake')
```

```python
data = [len(dataset_auth), len(dataset_fake)]
sns.barplot(x=['auth', 'fake'], y=data)
plt.title('Classes Count')
plt.ylabel('Frequency')
plt.show()
```

```python
path = './dataset/auth'
img = cv2.imread(f'{path}/{os.listdir(path)[15]}')

plt.imshow(img)
```

```python
# Specify the rotation angle (in degrees)
angle = 45

# Get the dimensions of the image
height, width = img.shape[:2]

# Calculate the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)

# Apply the rotation to the image
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

final_img = cv2.flip(rotated_image, 0)

plt.imshow(final_img)
```
